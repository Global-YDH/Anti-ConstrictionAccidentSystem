```python
import cv2
import numpy as np

# YOLO로 학습된 모델과 가중치 파일 로드
net = cv2.dnn.readNet('B:/ALMA_Detection/data.yaml', 'C:/Users/User/Desktop/TripleTrouble/yolov5-master/runs/train/result_jetbot2/weights/best.pt')

# 클래스 이름을 저장한 파일 로드
classes = []
with open('C:/Users/User/Desktop/TripleTrouble/yolov5-master/classes.txt', 'r') as f:
    classes = [line.strip() for line in f.readlines()]

# YOLO를 사용하여 문 객체 감지 수행하는 함수 정의
def detect_door_objects(image):
    height, width, _ = image.shape

    # 이미지 전처리
    blob = cv2.dnn.blobFromImage(image, 1/255.0, (416, 416), swapRB=True, crop=False)

    # 네트워크에 전달하여 객체 감지 수행
    net.setInput(blob)
    layer_outputs = net.forward(net.getUnconnectedOutLayersNames())

    # 감지된 객체 정보 추출
    boxes = []
    confidences = []
    class_ids = []
    for output in layer_outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            if confidence > 0.5 and class_id == 0:  # 클래스 ID가 0인 경우는 문 객체로 가정
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # 비최대 억제(NMS)를 적용하여 중복된 박스 제거
    indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    # 감지된 문 객체 정보 반환
    detected_objects = []
    for i in indices:
        i = i[0]
        box = boxes[i]
        confidence = confidences[i]
        class_id = class_ids[i]
        class_name = classes[class_id]
        detected_objects.append((box, confidence, class_id, class_name))

    return detected_objects

# 비디오 파일 열기
video_path = 'B:/data/KakaoTalk_20230511_162650048.mp4'
cap = cv2.VideoCapture(video_path)

# 비디오 프레임 읽기
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO를 사용하여 문 객체 감지 수행
    detected_objects = detect_door_objects(frame)

    # 문 객체 감지 결과를 이용한 추가 작업 수행
    for (box, confidence, class_id, class_name) in detected_objects:
        x, y, w, h = box
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, class_name, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 화면에 결과 출력
    cv2.imshow('Object Detection', frame)
    if cv2.waitKey(1) == ord('q'):
        break

# 비디오 파일 종료
cap.release()
cv2.destroyAllWindows()
```
